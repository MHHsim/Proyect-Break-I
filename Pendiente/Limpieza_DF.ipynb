```python
def cardinalidad(df_in, umbral_categoria, umbral_continua):

    resultados = []

    for col in df_in.columns:
        # Cardinalidad absoluta
        card = df_in[col].nunique(dropna=True)

        # Porcentaje de cardinalidad
        pct = card / len(df_in) * 100

        # Clasificación según reglas
        if card == 2:
            tipo = "Binaria"
        elif card < umbral_categoria:
            tipo = "Categorica"
        else:
            if pct >= umbral_continua:
                tipo = "Numerica Continua"
            else:
                tipo = "Numerica Discreta"

        # 👉 Guardamos el resultado de esta columna
        resultados.append({
            "columna": col,
            "cardinalidad": card,
            "pct_cardinalidad": pct,
            "tipo": tipo
        })

    return pd.DataFrame(resultados)
```

```python
def cardinalidad(df_in, umbral_categoria, umbral_continua):

    resultados = []

    for col in df_in.columns:
        serie = df_in[col]

        # 1. Tipo de dato
        if serie.dtype == "object":
            tipo = "Categorica"
        elif "datetime" in str(serie.dtype):
            tipo = "Fecha"
        else:
            # 2. Reglas de cardinalidad solo para numéricas
            card = serie.nunique(dropna=True)
            pct = card / len(df_in) * 100

            if card == 2:
                tipo = "Binaria"
            elif card < umbral_categoria:
                tipo = "Categorica"
            else:
                tipo = "Numerica Continua" if pct >= umbral_continua else "Numerica Discreta"

        resultados.append({
            "columna": col,
            "tipo": tipo
        })

    return pd.DataFrame(resultados)
```

```python
cardinalidad(df_2025, 12, 0.3)
```
