## HIPÓTESIS 1 — Los alojamientos ubicados en el Centro tienen precios más altos que los alojamientos del resto de barrios.

**1. Problema**
Queríamos comprobar si los alojamientos situados en el Centro tienen precios más altos que los del resto de barrios.
La idea inicial era que, al ser la zona más turística y con mayor demanda, sus precios deberían ser superiores.

**2. Datos**
Trabajamos con un conjunto de alojamientos turísticos que incluye información sobre:
- Precio por noche
- Barrio donde se ubica cada alojamiento
- Tipo de alojamiento
- Disponibilidad anual
- Número de reseñas

El Centro destaca porque concentra 6.389 alojamientos, lo que supone el 67 % de toda la oferta.

**3. Proceso**
Para evaluar la hipótesis seguimos un proceso claro y comprensible:
- Comparamos los precios del Centro con los del resto de barrios.
- Analizamos la distribución de tipos de alojamiento, ya que influyen directamente en el precio.
- Revisamos si la disponibilidad o el número de reseñas afectan al precio.
- Estudiamos los barrios más caros para entender por qué superan al Centro.

**4. Hallazgos**
Los datos muestran claramente que el Centro NO es el barrio más caro.
Estas son las razones principales:

- Exceso de oferta en el Centro: El Centro tiene más de 4.000 alojamientos, muy por encima del resto de barrios. Esta enorme oferta genera una fuerte competencia entre hosts, lo que tiende a moderar los precios.
Alta oferta → alta competencia → precios más ajustados.

- Predominan alojamientos más económicos
    En el Centro abundan:
        - 4.220 estudios completos (media: 123 €)
        - 368 habitaciones privadas (media: 56 €)
        - 8 habitaciones compartidas (media: 36 €)

El análisis demuestra que el tipo de alojamiento es el factor que más determina el precio.

- La disponibilidad no influye en el precio: La correlación entre disponibilidad y precio es prácticamente cero. Aunque el Centro tenga mucha rotación y demanda, la disponibilidad no empuja los precios hacia arriba.

- Los barrios más caros tienen menos oferta y alojamientos más grandes: Barrios como Campanillas, Churriana, Este o Puerto de la Torre tienen:
    - Menos alojamientos
    - Viviendas más grandes
    - Lo que suponemos que tiene una demanda más específica (playa, aeropuerto, PTA, zonas residenciales)
Esto eleva el precio medio de forma natural.

- El Centro es un mercado saturado y maduro
La saturación turística del Centro hace que los precios estén más estabilizados.
Los hosts ajustan tarifas para mantener ocupación en un entorno muy competitivo.

- Acciones
Dado que la hipótesis no se confirma, tomamos las siguientes decisiones:

- No usar la media como indicador principal en barrios con mucha variabilidad: La mediana y la distribución por tipo de alojamiento ofrecen una visión más realista.

- Analizar los factores que realmente explican el precio
    - Tipo de alojamiento
    - Número de reseñas
    - Disponibilidad anual

- Estudiar individualmente los barrios con precios anómalos
Especialmente aquellos con pocos alojamientos y valores extremos que distorsionan la media.

#### **Conclusión final**
**El Centro no es el barrio más caro porque su enorme oferta y la fuerte competencia mantienen los precios moderados.
Los barrios más caros lo son porque tienen menos alojamientos, viviendas más grandes y una demanda más específica que eleva el precio medio.**

## HIPÓTESIS 2 — Precios según el volumen de reseñas

**1. Problema**
Queríamos comprobar si el precio de los alojamientos con pocas reseñas es más alto que el de los que tienen muchas. La lógica detrás de esto es que los anfitriones novatos tienden a sobrevalorar su propiedad o carecen de precios optimizados, mientras que los veteranos ajustan tarifas para asegurar la ocupación.

**2. Datos**
Analizamos la relación entre las tarifas por noche y el volumen acumulado de reseñas de los alojamientos, segmentando los resultados según el tipo de propiedad (apartamentos completos, habitaciones privadas y habitaciones compartidas).

**3. Proceso**
- Evaluamos la distribución de precios en función de la veteranía del anuncio (número de reseñas).
- Cruzamos los datos con los diferentes tipos de alojamiento para verificar si el comportamiento del mercado era homogéneo.
- Identificamos desviaciones y patrones anómalos en los segmentos analizados.

**4. Hallazgos clave**
- **Comportamiento general:** Para el grueso del mercado (apartamentos completos y habitaciones privadas), el precio mediano desciende a medida que el alojamiento acumula más reseñas. Esto confirma que los nuevos anuncios suelen entrar con precios de prueba superiores.
- **La excepción del bajo coste:** En las habitaciones compartidas la tendencia se invierte por completo. Los alojamientos con más reseñas son más caros (50€) que los nuevos (34.5€). Aquí, la confianza y reputación que aportan las reseñas actúan como un activo de valor que permite subir el precio.
- **Efecto disponibilidad:** Los alojamientos con pocas reseñas muestran precios medios más altos debido a que suelen estar disponibles en fechas de máxima demanda, cuando las opciones más económicas ya han sido reservadas por completo.

#### **Conclusión final**
**La hipótesis se confirma para el grueso del mercado (casas enteras y habitaciones privadas), donde la falta de optimización o el posicionamiento inicial elevan los precios de los anuncios nuevos. Sin embargo, no es válida para el segmento "low-cost" de habitaciones compartidas, donde la veteranía y las reseñas positivas son el factor clave que permite a los hosts exigir tarifas más altas.**

## HIPÓTESIS 3 — Análisis de la demanda estacional por distritos

**1. Problema**
Buscábamos validar si la demanda turística en Málaga (medida a través del volumen de reseñas mensuales) experimenta un pico de crecimiento superior al 50% durante el tercer trimestre del año (julio, agosto y septiembre), y si el Distrito Centro actúa como la zona más estable y menos expuesta a esta estacionalidad veraniega.

**2. Datos**
Trabajamos combinando el histórico de opiniones mensuales (`reviews.csv`) con la localización geográfica de las propiedades (`listings.csv`), logrando asignar un barrio y distrito específico a cada una de las reseñas registradas para medir el impacto temporal.

**3. Proceso**
- Consolidamos y limpiamos las fechas de los registros históricos para agruparlas por trimestres.
- Realizamos un análisis porcentual de crecimiento comparando el tercer trimestre (Q3) frente al resto del año.
- Clasificamos los distritos de Málaga mediante un análisis multivariante en función de su comportamiento y estabilidad ante la temporada estival.

**4. Hallazgos clave**
- **Estabilidad absoluta del Centro:** El Distrito Centro presenta una variación porcentual mínima en verano (apenas un 0.29%). Esto demuestra que no sufre de picos estacionales porque opera de forma constante a su máxima capacidad operativa durante todo el año; es el motor incombustible de la ciudad.
- **Zonas de Alta Estacionalidad:** Distritos como Teatinos y Palma-Palmilla muestran picos de demanda sumamente agresivos concentrados de forma exclusiva en los meses de verano. Su viabilidad económica depende estrechamente de la temporada estival.
- **Zonas de Estabilidad Madura:** Carretera de Cádiz y Campanillas acompañan al Centro en esta categoría. Aunque registran un volumen alto de actividad en verano, mantienen métricas muy sólidas y una ocupación estable durante el resto de los meses del año.
- **Comportamiento anómalo:** Churriana destaca de forma negativa al registrar un crecimiento inverso (decrecimiento) durante el tercer trimestre, marcando una clara línea de investigación para entender este comportamiento atípico.

#### **Conclusión final**
**La hipótesis se cumple parcialmente: existe una estacionalidad extrema en la periferia, pero el Distrito Centro no experimenta un pico del 50% porque sus niveles de demanda ya son máximos de forma sostenida durante todo el año. Málaga se divide así en zonas de "alta estacionalidad" que dependen del verano y zonas de "estabilidad madura" capaces de atraer volumen constante de manera desestacionalizada.**
