## HIPÓTESIS 1 — Los alojamientos ubicados en el Centro tienen precios más altos que los alojamientos del resto de barrios.

### **1. Problema**
Queríamos comprobar si los alojamientos situados en el Centro tienen precios más altos que los del resto de barrios.
La idea inicial era que, al ser la zona más turística y con mayor demanda, sus precios deberían ser superiores.

### **2. Datos**
Trabajamos con un conjunto de alojamientos turísticos que incluye información sobre:
- Precio por noche
- Barrio donde se ubica cada alojamiento
- Tipo de alojamiento
- Disponibilidad anual
- Número de reseñas

El Centro destaca porque concentra 6.389 alojamientos, lo que supone el 67 % de toda la oferta.

### **3. Proceso**
Para evaluar la hipótesis seguimos un proceso claro y comprensible:
- Comparamos los precios del Centro con los del resto de barrios.
- Analizamos la distribución de tipos de alojamiento, ya que influyen directamente en el precio.
- Revisamos si la disponibilidad o el número de reseñas afectan al precio.
- Estudiamos los barrios más caros para entender por qué superan al Centro.

### **4. Hallazgos**
Los datos muestran claramente que el Centro NO es el barrio más caro.
Estas son las razones principales:

- Exceso de oferta en el Centro: El Centro tiene más de 4.000 alojamientos, muy por encima del resto de barrios. Esta enorme oferta genera una fuerte competencia entre hosts, lo que tiende a moderar los precios.
Alta oferta → alta competencia → precios más ajustados.

- Predominan alojamientos más económicos
    En el Centro abundan:
        - 4.220 estudios completos (media: 123 €)
        - 368 habitaciones privadas (media: 56 €)
        - 8 habitaciones compartidas (media: 36 €)

El análisis demuestra que el tipo de alojamiento es el factor que más determina el precio.

- La disponibilidad no influye en el precio: La correlación entre disponibilidad y precio es prácticamente cero. Aunque el Centro tenga mucha rotación y demanda, la disponibilidad no empuja los precios hacia arriba.

- Los barrios más caros tienen menos oferta y alojamientos más grandes: Barrios como Campanillas, Churriana, Este o Puerto de la Torre tienen:
    - Menos alojamientos
    - Viviendas más grandes
    - Lo que suponemos que tiene una demanda más específica (playa, aeropuerto, PTA, zonas residenciales)
Esto eleva el precio medio de forma natural.

- El Centro es un mercado saturado y maduro
La saturación turística del Centro hace que los precios estén más estabilizados.
Los hosts ajustan tarifas para mantener ocupación en un entorno muy competitivo.

- Acciones
Dado que la hipótesis no se confirma, tomamos las siguientes decisiones:

- No usar la media como indicador principal en barrios con mucha variabilidad: La mediana y la distribución por tipo de alojamiento ofrecen una visión más realista.

- Analizar los factores que realmente explican el precio
    - Tipo de alojamiento
    - Número de reseñas
    - Disponibilidad anual

- Estudiar individualmente los barrios con precios anómalos
Especialmente aquellos con pocos alojamientos y valores extremos que distorsionan la media.

#### **Conclusión final**
**El Centro no es el barrio más caro porque su enorme oferta y la fuerte competencia mantienen los precios moderados.
Los barrios más caros lo son porque tienen menos alojamientos, viviendas más grandes y una demanda más específica que eleva el precio medio.**

## HIPÓTESIS 2 — Relación inversa entre veteranía (reseñas) y precio de salida

### **1. Problema**
Queríamos comprobar si existe una relación inversa entre la veteranía de un alojamiento (medida por su volumen acumulado de opiniones) y su precio por noche. La intuición de negocio dictaba que los anfitriones novatos o propiedades nuevas tienden a sobrevalorar sus inmuebles por falta de optimización, mientras que los anfitriones consolidados ajustan sus precios a la baja para garantizar ocupación.

### **2. Datos**
Para el análisis cruzamos las siguientes variables de nuestro dataset limpio de 2025:
* `price`: Tarifa limpia por noche (sin símbolos ni outliers extravagantes).
* `number_of_reviews`: Historial total de reseñas como indicador de veteranía en la plataforma.
* `calculated_host_listings_count`: Volumen de propiedades por host, para evaluar el nivel de profesionalización.

### **3. Proceso**
* **Segmentación por madurez:** Clasificamos los alojamientos en tres categorías según su número de reseñas: "Novatos" (0-10 reseñas), "Intermedios" (11-50 reseñas) y "Veteranos" (más de 50 reseñas).
* **Análisis estadístico:** Calculamos el coeficiente de correlación de Pearson entre las dos variables y analizamos las medianas de precio de cada grupo para evitar el sesgo de las propiedades de gran lujo.
* **Visualización:** Creamos un gráfico de dispersión (Scatter Plot) combinado con diagramas de caja (Box Plots) para observar la dispersión del precio en cada grupo de veteranía.

### **4. Resultados**
* **Ausencia de correlación:** El análisis estadístico arrojó una correlación prácticamente nula (0.0), demostrando que el precio no disminuye a medida que aumentan las reseñas.
* **Homogeneidad en los precios:** Al observar las medianas de precios por grupos, descubrimos que los alojamientos "Novatos" y los "Veteranos" mantienen un precio mediano de salida sumamente plano y equilibrado en el mercado de Málaga. 
* **El factor profesionalización:** Detectamos que los precios más altos pertenecen a hosts que gestionan múltiples propiedades (perfil profesional), quienes fijan tarifas competitivas mediante herramientas de precios dinámicos desde el primer día que publican el anuncio.

#### **Conclusión final**
**La hipótesis SE RECHAZA.** Los datos demuestran que no existe una relación inversa entre veteranía y precio en el mercado de Málaga. Un anfitrión con más de 100 reseñas no abarata su propiedad. La fuerte presencia de gestores profesionales provoca que los precios se fijen siguiendo la lógica de la oferta de la zona y el tipo de inmueble, eliminando por completo el supuesto sesgo de "error de novato".

## HIPÓTESIS 3 — Estacionalidad de la demanda a nivel global y por distritos

### **1. Problema**
El objetivo era medir el impacto del factor temporal en el comportamiento del turismo en Málaga. Planteamos que, al ser un destino costero, la demanda sufriría una explosión agresiva en verano (Q3) con un crecimiento superior al 50% respecto a la primavera (Q2). Asimismo, sosteníamos que el Distrito Centro se comportaría como un refugio de estabilidad homogénea debido a su potente atractivo cultural y gastronómico independiente del clima.

### **2. Datos**
Conectamos un volumen masivo de datos uniendo dos fuentes de información mediante sus identificadores únicos (`listing_id` / `id`):
* El histórico reciente de opiniones de clientes (**197.476 reseñas** de los años 2024 y 2025) como métrica directa de demanda real/ocupación.
* La columna `neighbourhood` de nuestro dataset limpio para geolocalizar la actividad.

### **3. Proceso**
* **Tratamiento temporal:** Transformamos la columna de fechas a formato temporal, extrayendo el mes analítico (1 al 12) y el trimestre del año (Q1 a Q4).
* **Cruce y normalización:** Cruzamos las opiniones con los barrios. Dado que distritos como el Centro concentran la gran mayoría de la oferta y "aplastarían" visualmente a los demás, aplicamos una **normalización porcentual**. Calculamos el porcentaje interno de reservas que recibe cada barrio en cada trimestre sobre su propio 100% anual, garantizando una comparación justa.
* **Visualización:** Dibujamos un análisis bivariante mensual global (gráfico de barras por meses) y un gráfico multivariante de barras agrupadas por distritos y trimestres.

### **4. Resultados**
* **El fenómeno de la Desestacionalización Global:** Rompiendo con la intuición inicial, la variación de la demanda global en verano (Q3) frente a la primavera (Q2) fue de apenas un **-0.66%**. La curva es totalmente plana entre abril y agosto, con picos idénticos en mayo y agosto.
* **Estabilidad absoluta en el Centro:** La hipótesis sobre el corazón de la ciudad se validó por completo. El Distrito Centro mostró una variación veraniega insignificante del **-3.88%**, manteniendo una distribución homogénea excelente (32.42% en primavera y 31.16% en verano). Su oferta cultural lo inmuniza contra la dependencia estival.
* **La anomalía radical de Teatinos-Universidad:** El análisis multivariante destapó un patrón oculto espectacular en Teatinos. Mientras la ciudad se mantiene plana, la demanda en Teatinos se dispara un **+124.47%** en verano, llegando a concentrar de golpe el **57.49%** de todas sus reservas de todo el año en el tercer trimestre (Q3).

#### **Conclusión final**
**La hipótesis SE CUMPLE PARCIALMENTE.** Se rechaza de forma contundente en el plano global, demostrando que Málaga ha logrado un éxito absoluto en la desestacionalización de su turismo, adelantando la temporada alta a primavera. Sin embargo, se acepta firmemente en el ámbito geográfico: el Centro es el motor más estable del año, mientras que la periferia estudiantil (Teatinos) sufre una estacionalidad extrema y agresiva ligada a la liberación de pisos universitarios para uso turístico en vacaciones.
