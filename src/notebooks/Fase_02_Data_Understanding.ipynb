## 2. Inspección Técnica de Variables (Data Dictionary)
Analizamos los tipos de datos asignados automáticamente por Python y detectamos la presencia de valores nulos iniciales.

```python
# Generación del data dictionary automático
data_dict = pd.DataFrame({
    'columna': df.columns,
    'dtype': df.dtypes.values,
    'nulos_pct': (df.isnull().mean() * 100).round(1).values,
    'unicos': df.nunique().values,
    'ejemplo': [df[c].dropna().iloc[0] if len(df[c].dropna()) > 0 else None for c in df.columns]
})

display(data_dict)
```

```python
#  # 1. Creamos diccionarios sencillos para las descripciones de manera manual y y la clasificación estadística según sea Numérica (Discreta o Continua) 
# o Categórica (Nominal/Ordinal).
descripciones_dict = {
    'id': 'Identificador único del anuncio',
    'name': 'Nombre comercial del alojamiento',
    'host_id': 'Identificador único del anfitrión',
    'host_name': 'Nombre del anfitrión',
    'neighbourhood_group': 'Zona o distrito principal (agrupación de barrios)',
    'neighbourhood': 'Barrio específico donde se ubica el inmueble',
    'latitude': 'Coordenada geográfica de latitud',
    'longitude': 'Coordenada geográfica de longitud',
    'room_type': 'Tipo de alojamiento (Piso completo, Habitación, etc.)',
    'price': 'Precio por noche en moneda local',
    'minimum_nights': 'Número mínimo de noches para reservar',
    'number_of_reviews': 'Total de reseñas acumuladas',
    'last_review': 'Fecha en la que se publicó la última reseña',
    'reviews_per_month': 'Promedio de reseñas recibidas mensualmente',
    'calculated_host_listings_count': 'Total de propiedades que posee el mismo host',
    'availability_365': 'Días disponibles para reserva en el año',
    'number_of_reviews_ltm': 'Reseñas recibidas en los últimos 12 meses',
    'license': 'Número de licencia turística obligatoria'
}

tipo_dict = {
    'id': 'Categórica Nominal (ID)',
    'name': 'Categórica Nominal',
    'host_id': 'Categórica Nominal (ID)',
    'host_name': 'Categórica Nominal',
    'neighbourhood_group': 'Categórica Nominal',
    'neighbourhood': 'Categórica Nominal',
    'latitude': 'Numérica Continua',
    'longitude': 'Numérica Continua',
    'room_type': 'Categórica Nominal',
    'price': 'Numérica Continua',
    'minimum_nights': 'Numérica Discreta',
    'number_of_reviews': 'Numérica Discreta',
    'last_review': 'Temporal',
    'reviews_per_month': 'Numérica Continua',
    'calculated_host_listings_count': 'Numérica Discreta',
    'availability_365': 'Numérica Discreta',
    'number_of_reviews_ltm': 'Numérica Discreta',
    'license': 'Categórica Nominal'
}

# 2. Mapeamos directamente usando los diccionarios que hemos creado antes, con el método .map()
data_dict['descripcion'] = data_dict['columna'].map(descripciones_dict)
data_dict['clasificacion'] = data_dict['columna'].map(tipo_dict)

# 3. Reordenamos para que la descripción quede al lado del nombre de la columna, y mostramos
data_dict = data_dict[['columna', 'descripcion', 'clasificacion_estadistica', 'dtype', 'nulos_pct', 'unicos', 'ejemplo']]

display(data_dict)
```
